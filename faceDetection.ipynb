A simple implementation of code for detecting faces in a given image

```python
# Import Librairies
import dlib
import cv2

# Google colab imports
from google.colab.patches import cv2_imshow

# Access to google drive
from google.colab import drive
drive.mount('/content/drive')
```

# Detecting facial points

```python
face_detector = dlib.get_frontal_face_detector()
# Each face will have 68 points, in case of a mobile phone we should 5 face landmark file
points_detector = dlib.shape_predictor('/content/drive/MyDrive/computer vision/weights/shape_predictor_68_face_landmarks.dat')
```

```python
image = cv2.imread('/content/drive/MyDrive/computer vision/people.jpg')
face_detection = face_detector(image, 1)
for face in face_detection:
  points = points_detector(image, face)
  print(len(points.parts())) # Proving that we get 68 points needed to detect the face
  for point in points.parts():
    cv2.circle(image, (point.x, point.y), 2, (0, 255, 0), 2)
  l, t, r, b = face.left(), face.top(), face.right(), face.bottom()
  cv2.rectangle(image, (l, t), (r, b), (0, 255, 255), 2)
cv2_imshow(image)
```
